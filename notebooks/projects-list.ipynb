```python
import pandas as pd
import os
import sys
```

```python
sys.path.insert(0,'../code')
import objects
```

```python
path_raw = '../data/raw/projects/'
```

```python
df = pd.DataFrame()
for file in os.listdir(path_raw):
    
    if file.endswith('.xlsx'):
        df_temp = pd.read_excel(
            os.path.join(path_raw, file),
            skiprows=1,
            engine='openpyxl'
        )
        df = pd.concat([df, df_temp]).reset_index(drop=True)
```
