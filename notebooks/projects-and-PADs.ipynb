```python
def is_pad(string):
    
    id_pattern = re.compile(r'PROJECT(?: |\xa0)+APPRAISAL(?: |\xa0)+DOCUMENT')
    result = id_pattern.search(string)
    
    if result:
        return True
    else:
        return False
```

```python
df_pad_contents['is_pad'] = df_pad_contents['pad_content'].apply(is_pad)
```

```python
def clean(string):
    
    string = re.sub('\n.*[pP]age [0-9i]{1,3} of [0-9]{0,3}', '\n', string)
    string = re.sub('\xa0', ' ', string)
    string = re.sub('\x0c +The World Bank', '\n', string)
    string = re.sub('\n\s+\n', '\n\n', string)
    string = re.sub('\n+', '\n', string)
    string = string.lower()
    
    return string
```

```python
def components(string):
    
    pattern = re.compile('component name\s+cost[\s\S]+$')
    string2 = pattern.search(string)
    
    if string2:
        lines = string2.group(0).splitlines()[1:]
        comps = []
        counter = 0
        limit = 3
        skip = []
        
        for i, line in enumerate(lines):
            
            line = line.strip()
            
            if line == 'organizations':
                break
            if i in skip:
                continue
                
            pattern = re.compile(r'^\S[\s\S]+\S\s+(?:(?:us\$)|(?:usd)|(?:\$)){0,1}[0-9,.]{0,13}[0-9]$')
            match = pattern.search(line)

            if match:
                comps.append(match.group(0))
                counter = 0
            else:
                match2 = pattern.search(lines[i+1].strip())
                match3 = pattern.search(lines[i+2].strip())
                
                if not match2 and not match3:
                
                    line = line + ' ' + lines[i+2].strip() + lines[i+1]
                    match = pattern.search(line)

                    if match:
                        comps.append(match.group(0))
                        counter = 0
                        skip += [i+1, i+2]
                    else:                
                        counter += 1
                        if counter >= limit:
                            break
        
        return comps
        
    else:
        return []
```

```python
df_pad_contents['components'] = df_pad_contents['pad_content'].apply(lambda x: components(clean(x)))
```

```python
def split_component(string):
    
    name_pattern = re.compile(r'(^\S[\s\S]+\S)\s+')
    name = name_pattern.search(string).group(1)
    
    amount_pattern = re.compile(r'(?:(?:us\$)|(?:usd)|(?:\$)){0,1}[0-9,.]{0,13}[0-9]$')
    amount = amount_pattern.search(string).group(0)
    
    return name, amount
```

```python
df_components = pd.DataFrame(columns=['doi', 'component', 'raw_amount'])
df_temp = df_pad_contents[df_pad_contents['components'].astype(bool)]
bad_names = [
    'total',
    'vat',
    'sub-total',
    'subtotal',
    'sub-total without vat',
    'page',
    'ida'
]

for i in range(len(df_temp)):
    
    doi = df_temp['doi'].iloc[i]
    raw_components = df_temp['components'].iloc[i]
    for comp in raw_components:
        name, amount = split_component(comp)
        if name in bad_names:
            continue
        if amount == '2020':
            continue
        df_components.loc[len(df_components)] = [doi, name, amount]
```

```python
def clean_amount(string):
    
    # manually correcting one exception:
    if string == '289.89.00':
        string = '289.89'
    
    pattern = re.compile(r',[0-9]{2}$')
    match = pattern.search(string)
    if match:
        string_list = list(string)
        string_list[-3] = '.'
        string = ''.join(string_list)
    
    string = re.sub('[^0-9.]', '', string)
    amount = float(string)    
    
    if amount >= 1_000_000:
        amount = amount / 1_000_000
    
    
    return amount
```

```python
def clean_amount2(string):
    
    # manually correcting one exception:
    if string == '289.89.00':
        string = '289.89'
    
    pattern = re.compile(r',[0-9]{2}$')
    match = pattern.search(string)
    if match:
        string_list = list(string)
        string_list[-3] = '.'
        string = ''.join(string_list)
    
    string = re.sub('[^0-9.]', '', string)
    amount = float(string)    
    
    return amount
```

```python
df_components['amount'] = df_components['raw_amount'].apply(clean_amount)
df_components['amount_for_percentage'] = df_components['raw_amount'].apply(clean_amount2)
```

# Results

- total projects in 2012-2021
